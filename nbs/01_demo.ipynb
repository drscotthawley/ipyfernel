The `__init__.py` needs to explicitly re-export symbols from submodules for them to be available at the top level.

Quick diagnostic â€” try this:

```python
from ipyfernel.core import register_remote_kernel
```

If that works, then the issue is that `ipyfernel/__init__.py` isn't importing/re-exporting from `core.py`.
